# 01: Retrieval Baselines
Comparing Lexical (BM25) vs Dense vs Hybrid pipelines.

```python
from irlib import SearchPipeline, BM25Retriever, DenseRetriever
pipe = SearchPipeline(lexical_retriever=BM25Retriever(), dense_retriever=DenseRetriever())
pipe.index(['Legal document test 1', 'Search tutorial 2'])
print(pipe.search('legal search'))
```
